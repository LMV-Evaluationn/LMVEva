```python
# Comparación entre Honduras y Nicaragua #Distribución de Emisiones
```

```python
from matplotlib import pyplot as plt
```

```python
honduras = DatosTea [DatosTea.Área == "Honduras"]
nicaragua = DatosTea [DatosTea.Área == "Nicaragua"]
```

```python
plt.plot (honduras.Año, honduras.Valor, color="green", linewidth=4, linestyle="-")
plt.plot (nicaragua.Año, nicaragua.Valor,color="orange", linewidth=4, linestyle="-")
plt.xlabel ("Años",fontsize=13)
plt.ylabel ("valores",fontsize=13)
plt.title("Distribución de Emisiones",fontsize=15,color="red")
plt.legend(["Honduras","Nicaragua"])
plt.show()
```
